```python
import geopandas as gpd
import mapclassify
import libpysal
import seaborn as sns
sns.set_theme(style='dark')
```

## Liste de tous les exemples dipo sur libpysal

```python
libpysal.examples.available()
```

```python
_ = libpysal.examples.load_example('South')
pth = libpysal.examples.get_path('south.shp')
```

```python
south = gpd.read_file(pth)
south
```

```python
ax = south.plot(column='HR60', scheme='QUANTILES', k=4, \
             cmap='BuPu', legend=True,
             legend_kwds={'loc': 'center left', 'bbox_to_anchor':(1,0.5)})
```

On prend la lÃ©gende

```python
labels = [t.get_text() for t in ax.get_legend().get_texts()]
labels
```

On tire les quantiles extrait par mapclassify

```python
q4 = mapclassify.Quantiles(south.HR60, k=4)
q4
```

```python
labels == q4.get_legend_classes()
```

### Chqnger le format numÃ©rique de la lÃ©gende

```python
ax = south.plot(column='HR60', scheme='QUANTILES', k=4, \
             cmap='BuPu', legend=True,
             legend_kwds={'loc': 'center left', 'bbox_to_anchor':(1,0.5),  'fmt':"{:.4f}"}) # C'est grace au fmt
```

```python
ax = south.plot(column='HR60', scheme='QUANTILES', k=4, \
             cmap='BuPu', legend=True,
             legend_kwds={'loc': 'center left', 'bbox_to_anchor':(1,0.5),  'fmt':"{:.0f}"})
```

### Pour voir les crochets des intervalles

```python
ax = south.plot(column='HR60', scheme='BoxPlot', \
             cmap='BuPu', legend=True,
             legend_kwds={'loc': 'center left', 'bbox_to_anchor':(1,0.5),
                         'interval': True}) # parametre interval
```
